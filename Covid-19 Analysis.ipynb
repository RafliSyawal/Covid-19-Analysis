Berikut merupakan _leading insights_ dari analisis kami:

- **Prospek pemulihan COVID-19** \\
Baik untuk Indonesia dan Jakarta, masih terdapat ketidakpastian yang tinggi akan kasus positif-kematian COVID-19, walaupun fluktuasi sudah turun dibandingkan awal-awal terkena virus pada 2020. Dua bulan terakhir ada penurunan kasus dan angka ini stabil, akan tetapi, kami memprediksi dua jalur kemungkinan, kondisi realistis meningkat (data max yang selalu di atas Q3, prediksi LSTM) dan kondisi ideal yang tetap stabil (analisis VAR). Meningkat jika lengah, karena sudah ada pola-pola kasus rendah disusul dengan pertumbuhan eksponensial. Stabil atau meningkat terkendali jika mulai mengerem.

- **Vaksinasi**, dengan Jakarta sebagai studi kasus \\

Hanya saja banyak wilayah tertinggal, seperti ada yang masih 55.78% dan ada yang mencapai 784.86%. Dampaknya terhadap kasus positif tidak signifikan. Ada _rebound effect_ (makin banyak vaksin, lengah karena merasa kuat, jadi makin banyak mobilitas, berdampak pada peningkatan kasus positif), tetapi jumlah pasien sembuh semakin cepat meningkat juga.

- **Kebijakan yang dibutuhkan** \\
Kementerian / dinas kesehatan perlu bergeser fokus berdasarkan situasi kasus. Di saat kasus positif tinggi, fokus pada _treatment_ agar cepat pasien positif cepat recover. Di saat kasus rendah, fokus pada _testing-tracing_. Kampanye pencegahan dan vaksinasi harus terus berjalan. Pemerintah (eksekutif-legislatif) perlu metapkan indikator pasti kapan untuk mengerem-gas pembatasan mobilitas, untuk mencegah _build-up_ kasus pola eksponensial yang membebani infrastruktur kesehatan.

 \\

Di lain sisi, periode pertumbuhan rendah terjadi karena nilai lampau yang sudah tinggi: misalnya, dari 100 menjadi 110 kasus, hanya akan tercatat sebagai peningkatan 10%. \\
 \\
Periode pertumbuhan rendah menyiratkan ada banyaknya pasien positif dan penambahan kasus yang besar, sekalipun tertutupi oleh nilai lampau yang tinggi. Otomatis, prioritas infrastruktur kesehatan adalah untuk mengurangi jumlah orang yang terinfeksi. Pada periode pertumbuhan tinggi, jumlah kasus positif atau kematian serta pertambahannya masih rendah, sehingga fokus pembuat kebijakan harus berada pada tracing dan testing. \\
 \\
Ini juga terlihat dalam pola data di atas: dua ratus hari pertama COVID-19, di mana jumlah kasus masih sedikit di Indonesia, mencatat pertumbuhan tinggi, begitu pula dengan pengalaman terkini. ### Perkembangan Positif harian / Positif Aktif Jakarta, saat Sebelum dan Sesudah Vaksinasi

Dengan tersedianya vaksinasi, beban infrastruktur kesehatan seharusnya akan berkurang. Pasien terkena COVID-19 akan memiliki peluang selamat yang lebih tinggi. Bagaimanakah dinamikanya di Jakarta?

```python
df_covid_jkt['Hari Ke-'] = np.arange(1, len(df_covid_jkt)+1, 1)
vac_start_date = datetime.strptime('13/1/21', '%d/%m/%y')  # 13 Jan 2021
awal_vaksin = df_covid_jkt['Hari Ke-'][df_covid_jkt.tanggal==vac_start_date].values[0]-1

vac_boost_start_date = datetime.strptime('12/1/22', '%d/%m/%y') # 12 Januari 2022 awal vaksin booster
awal_vaksin_booster = df_covid_jkt['Hari Ke-'][df_covid_jkt.tanggal==vac_boost_start_date].values[0]-1

df_covid_jkt['PH/PA'] = df_covid_jkt['positif_harian'] / df_covid_jkt['positif_aktif']  # Positif Harian Relatif
df_covid_jkt['MH/PA'] = df_covid_jkt['meninggal_harian'] / df_covid_jkt['positif_aktif']  # Meninggal Harian Relatif
df_covid_jkt['SH/PA'] = df_covid_jkt['sembuh_harian'] / df_covid_jkt['positif_aktif']  # Sembuh Harian Relatif
```

```python
df_covid_jkt.info()
```

```python
df_covid_jkt = df_covid_jkt.dropna()
df_covid_jkt.tail()
```

```python
print('Perkembangan Positif Harian/Positif Aktif Sebelum 13 Jan 21')
print(df_covid_jkt['PH/PA'].iloc[:awal_vaksin].describe())
print()

print('Perkembangan Positif Harian/Positif Aktif Setelah 13 Jan 21')
print(df_covid_jkt['PH/PA'].iloc[awal_vaksin:].describe())
print()

print('Perkembangan Kasus Positif Harian Saat Vaksinasi Booster Diadakan')
print(df_covid_jkt['PH/PA'].iloc[awal_vaksin_booster:].describe())
print()
```

```python
tanggal = df_covid_jkt['tanggal']

fig, (ax1) = plt.subplots(1,1,figsize=(14,5))

ax1.scatter(x = tanggal,  y = df_covid_jkt['PH/PA'])

ax1.axhline(y=0.093099, xmin = (awal_vaksin-300)/497, xmax=(awal_vaksin-125)/497, c='r', 
            label='Mean sebelum vaksin (13 Jan 2021)')
ax1.axhline(y=0.125666, xmin = (awal_vaksin-125)/497, xmax=0.7858, c='r', 
            label='Mean setelah vaksin diadakan (13 Jan 2021)', linestyle = 'dashed')

ax1.axhline(y=0.132040, xmin = (awal_vaksin_booster-100)/497, xmax=0.781, c='y', 
            label='Mean saat vaksin booster diadakan')

ax1.axvline(x = vac_start_date, color = 'black', linestyle = ':', label="Vaksin 1 & 2")

ax1.axvline(x = vac_boost_start_date, color = 'green', linestyle = ':', label="Vaksin Booster")


plt.title('Positif Harian')
plt.grid(False)
plt.legend()
```

Gambaran statistik deskriptif menunjukkan bahwa terdapat peningkatan nilai mean dengan sebelum 13 Januari 2021 senilai 0.093 menjadi 0.125666 setelah 13 Januari 2021 dan juga terjadi peningkatan kasus positif pada tahun 2022 menjadi 0.132040. Meskipun demikian, grafik menunjukkan bahwa positif harian relatif terhadap banyaknya kasus yang sedang aktif setiap waktunya cenderung stabil dan stagnan. Peningkatan yang terjadi tidak signifikan karena perubahan rerata pada periode-periode yang kami buat, belum sampai melewati simpangan baku. \\
 \\
Indikator ini mengukur tingkat positif harian dengan kasus positif yang masih aktif di DKI Jakarta, sepanjang waktu. Peningkatan ini tidak signifikan secara statistik, karena jika dilihat dari nilai mean sebuah positif harian terhadap positif aktif bahwa ditemukkan terdapat peningkatan nilai tersebut namun tidak terlalu signifikan . Hal yang perlu diperhatikan adalah karena keterbatasan data yang tersedia, **analisis ini hanya menunjukkan kondisi keseluruhan pandemi di DKI Jakarta antara sebelum dan sesudah vaksinasi berjalan, bukan mengukur efektivitas dari vaksin itu sendiri.**

```python
fig, ax1 = plt.subplots()
(df_covid_jkt['positif_harian'].rolling(window=7).std()/df_covid_jkt['positif_aktif'].rolling(window=7).std()).plot(figsize=(12,6))
(df_covid_jkt['positif_harian'].rolling(window=30).std()/df_covid_jkt['positif_aktif'].rolling(window=7).std()).plot(figsize=(12,6))
ax.legend(["Simp. baku 7 hari", "Simp. baku 30 hari"])
ax.set_title("Fluktuasi dan ketidakpastian relatif, positif harian terhadap positif aktif")
```

##### Import Library

```python
import datetime
import operator
import matplotlib.pyplot as plt
from ipywidgets import interact
from IPython.display import Markdown as md
from datetime import date, timedelta, datetime
from numpy import sqrt
from pandas import read_csv, to_datetime
from prettytable import PrettyTable
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

##### Import Data

```python
df_indo = read_csv('/content/Covid-19(April-Juni2022).csv')
df_indo['indeks'] = df_indo.index
df_indo.reset_index(drop=True)
```

##### Preprocessing Data

```python
df_indo = df_indo.loc[:, ['indeks','date', 'daily_positive','cum_positive']]
# df['indeks'] = df.index
df_indo.columns = ['indeks', 'tanggal', 'positif_harian','akumulasi_positif']
df_indo
```

```python
degree = 2

x = df_indo['indeks'].values.reshape(df_indo['indeks'].size,1)
y = df_indo['positif_harian'].values.reshape(df_indo['positif_harian'].size,1)
polynomial_features= PolynomialFeatures(degree=degree)
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

rmse = sqrt(mean_squared_error(y,y_poly_pred))
r2 = r2_score(y,y_poly_pred)
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import sys
from datetime import timedelta
import tensorflow.keras as keras
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
np.set_printoptions(threshold=sys.maxsize)
```

### Data Preprocessing and Visualization

```python
# Add moving average (7 days) column
cases_ID_MA = cases_ID.copy()
cases_JKT_MA = cases_JKT.copy()

def AddMAForEachColumn(data: pd.DataFrame, MA_len: int):
    for col in data.columns:
        data[f'{col}_MA7'] = data[col].rolling(MA_len).mean()
    data = data.fillna(0)
    return data

cases_ID_MA = AddMAForEachColumn(cases_ID_MA, 7)
cases_JKT_MA = AddMAForEachColumn(cases_JKT_MA, 7)
```

```python
# Hyperparameters
history_len = 50
predict_len = 1
batch_size = 1000
shuffle_buffer_size = 256
train_split = 0.8
seed = 20
```

```python
# Data normalization
data_ID = cases_ID_MA[['daily_positive_MA7']].copy()
data_JKT = cases_JKT_MA[['daily_positive_MA7']].copy()

case_scaler_ID = MinMaxScaler()
data_ID_norm = pd.DataFrame(case_scaler_ID.fit_transform(data_ID), index=data_ID.index, columns=data_ID.columns)

case_scaler_JKT = MinMaxScaler()
data_JKT_norm = pd.DataFrame(case_scaler_JKT.fit_transform(data_JKT), index=data_JKT.index, columns=data_JKT.columns)
```

```python
# Data windowing
def get_windowed_data(data: pd.DataFrame, history_len: int, predict_len: int, columns: list):
    data_value = data[columns].copy().values
    xs = []
    ys = []
    for i in range(len(data_value[:-(history_len+predict_len-1), :])):
        xs.append(data_value[i:i+history_len, :])
        ys.append(data_value[i+history_len:i+history_len+predict_len,:])
    xs = np.array(xs)
    ys = np.array(ys)
    return xs, ys
  
xs_ID, ys_ID = get_windowed_data(data_ID_norm, history_len, predict_len, ['daily_positive_MA7'])
xs_ID = xs_ID.squeeze()
ys_ID = ys_ID.squeeze()

xs_JKT, ys_JKT = get_windowed_data(data_JKT_norm, history_len, predict_len, ['daily_positive_MA7'])
xs_JKT = xs_JKT.squeeze()
ys_JKT = ys_JKT.squeeze()
```

```python
# Data splitting
def split_data(xs: np.ndarray, ys: np.ndarray, train_split: int):
    # data is <class 'pandas.core.frame.DataFrame'>
    split = int(train_split * xs.shape[0])
    x_train, x_val = xs[:split], xs[split:]
    y_train, y_val = ys[:split], ys[split:]
    # train and val is <class 'pandas.core.frame.DataFrame'>
    return x_train, y_train, x_val, y_val

x_train_ID, y_train_ID, x_val_ID, y_val_ID = split_data(xs_ID, ys_ID, train_split)
x_train_JKT, y_train_JKT, x_val_JKT, y_val_JKT = split_data(xs_JKT, ys_JKT, train_split)
```

### Modelling

```python
num_of_features = 1
```

```python
tf.random.set_seed(seed)
model_LSTM_ID = keras.models.Sequential([
    keras.layers.LSTM(64, return_sequences=True, input_shape=(x_train_ID.shape[1], num_of_features)),
    keras.layers.Dropout(0.2),
    keras.layers.LSTM(128, return_sequences=True),
    keras.layers.Dropout(0.2),
    keras.layers.LSTM(64, return_sequences=True),
    keras.layers.Dropout(0.2),
    keras.layers.LSTM(64),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(predict_len),
])
model_LSTM_ID.compile(optimizer='adam', loss='mse', metrics=['mse', 'mae'])
model_LSTM_JKT = keras.models.Sequential([
    keras.layers.LSTM(64, return_sequences=True, input_shape=(x_train_JKT.shape[1], num_of_features)),
    keras.layers.Dropout(0.2),
    keras.layers.LSTM(128, return_sequences=True),
    keras.layers.Dropout(0.2),
    keras.layers.LSTM(64, return_sequences=True),
    keras.layers.Dropout(0.2),
    keras.layers.LSTM(64),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(predict_len),
])
model_LSTM_JKT.compile(optimizer='adam', loss='mse', metrics=['mse', 'mae'])
```

```python
history_ID = model_LSTM_ID.fit(x_train_ID, y_train_ID, epochs=80, batch_size=32)
history_JKT = model_LSTM_JKT.fit(x_train_JKT, y_train_JKT, epochs=80, batch_size=32)
```

Berdasarkan analisis statistik sederhana yang telah dilakukan menggunakan dataset dari situs resmi pemantauan Covid-19 Pemprov DKI Jakarta corona.jakarta.go.id, diperoleh hasil sebagai berikut:

- Namun relatif terhadap keparahan kondisi pandemi (‘Positif Aktif’) setiap waktunya, diperoleh tren penurunan laju penyebaran (positif harian) dan fatalitas (meninggal harian), sedangkan rata-rata sembuh harian mengalami peningkatan meskipun dalam tren penurunan setelah tanggal 13 Januari 2021. Artinya, relatif terhadap tingkat virality wabah Covid-19 di DKI Jakarta, diperoleh kondisi yang lebih baik setelah program vaksinasi 13 Januari 2021 dilaksanakan.

secara garis besar, kasus sembuh harian semakin meningkat secara fluktuatif dibandingkan kasus meninggal harian yang stabil dan kasus positif harian yang relatif stabil. Apabila hal ini terus dipertahankan tentunya merupakan hal yang baik dan menandakan peran dari fasilitas kesehatan yang mengerjakan tugas dengan baik dalam merawat pasien-pasien COVID-19. 

Dalam kajian ini, belum ditelusuri secara lebih mendalam mengenai kaitan antara tren jumlah positif harian dan jumlah sembuh harian. Ada beberapa faktor yang mungkin memberikan kontribusi terhadap hal ini. Misalkan saja, jumlah positif harian yang cenderung stabil diakibatkan oleh penerapan pembatasan sosial (PSBB) yang dapat menghambat laju penambahan kasus baru. Jumlah pasien sembuh yang semakin meningkat bisa disebabkan oleh instansi, sarana, dan prasarana kesehatan di DKI Jakarta yang sudah semakin belajar dan tanggap dalam menangani pasien COVID-19.

Positivity rate adalah perbandingan antara orang yang dinyatakan positif COVID-19 dengan jumlah tes yang dilakukan. Hampir sama dengan pola di seluruh provinsi lain di Indonesia, banyaknya jumlah test baru meningkat secara signifikan di bulan April 2020, kemudian Juni 2020. Ada berbagai persiapan sarana dan prasarana yang diperlukan untuk dapat melakukan tes secara masif dan sistematis. Positivity rate untuk Jakarta melonjak dan berfluktuasi cukup tajam di bulan Maret 2020 sampai pertengahan April 2020, mencapai >50%. Positivity rate kemudian menurun dan relatif stabil <10% semenjak pertengahan April 2020 sampai awal tahun 2021, dan terjadi peningkatan kembali di pertengahan bulan Juli 2021. Lalu upaya yang dilakukan oleh pemerintahan DKI-Jakarta dalam menangani kasus covid-19 dengan diadakan nya PPKM dapat memukul laju penurunan kasus positif covid-19 hingga tahun 2022, meskipun di awal tahun 2022 terjadi peningkatan kasus covid-19 namun jumlah positif kasus covid tersebut tidak separah dengan tahun sebelumnya.


Salah satu kesimpulan yang bisa diambil dari grafik adalah kapasitas testing yang dilakukan DKI Jakarta merupakan hal yang baik dan terus ditingkatkan sehingga akhirnya mampu memenuhi standar yang ditentukan WHO. Referensi standar WHO positivity rate < 5%. Positivity rate menjadi indikator yang dijadikan dasar apakah persebaran COVID-19 berkurang atau tidak, bukan berdasarkan jumlah kasus positif. Dengan memperhatikan hal ini, setiap daerah didorong untuk meningkatkan kapasitas testing.

Meskipun saat ini, sudah memasuki fase "New Normal", masyarakat DKI Jakarta harus
tetap sadar dan waspada terhadap COVID-19. Salah satu karakter suatu pandemi adalah dapat
mempercepat, memperlambat, dan mengulang kejadian itu sendiri, sehingga pemerintah dan
masyarakat perlu berhati-hati akan kemungkinan terjadinya gelombang berikutnya.

Kajian ini tentunya masih memerlukan pengembangan lebih jauh, terutama dalam melihat
keterkaitan variabel mobilitas dan variabel persebaran COVID-19 secara lebih sistematis dan
mendalam. Perhitungan korelasi dapat dikerjakan dengan lebih detail di waktu-waktu ke depan.
Dalam konteks lingkungan nyata, ada sangat banyak faktor yang dapat memberikan dampak dari
suatu inisiatif/ kebijakan yang diambil. Dalam kondisi ini, diperlukan cara analisa yang dapat
mempertimbangkan faktor-faktor yang begitu banyak, sebelum menyatakan kesimpulan dari
dampak suatu kebijakan

Salah satu contoh teknik yang sangat mungkin untuk dilakukan
adalah causal impact analysis. Dengan teknik ini, dapat dipastikan bahwa tren/ dampak yang
terjadi adalah benar-benar diakibatkan oleh variabel yang sedang diamati, bukan karena
faktor/variabel lain yang sangat mungkin memberikan pengaruh terhadap variabel yang sedang
diamati.
